```python
import  numpy as np
import matplotlib.pyplot as plt
import sympy
import math
```

```python
t0, t1, t2 = sympy.symbols("t0, t1, t2")
Funcion_Costo = (-sympy.log(1.0-(1.0/(1.0+sympy.exp(-1.0*(t0+t1*1.0+t2*2.0))))) -sympy.log(1-(1/(1+sympy.exp(-1*(t0+t1*3+t2*3)))))-sympy.log(1/(1+sympy.exp(-1*(t0+t1*4+t2*9)))))*1/3
Funcion_Costo
```

```python
Derivada_Funcion_Costo = [Funcion_Costo.diff(t0) , Funcion_Costo.diff(t1), Funcion_Costo.diff(t2)]
Derivada_Funcion_Costo
```

```python
Gradiente = sympy.Matrix(Derivada_Funcion_Costo)
Gradiente
```

```python
Hesseiana = [[Funcion_Costo.diff(t0,t0), Funcion_Costo.diff(t0,t1), Funcion_Costo.diff(t0,t2)], 
             [Funcion_Costo.diff(t1,t0), Funcion_Costo.diff(t1,t1), Funcion_Costo.diff(t1,t2)],
             [Funcion_Costo.diff(t2,t0), Funcion_Costo.diff(t2,t1), Funcion_Costo.diff(t2,t2)]]
Hesseianaa = sympy.Matrix(Hesseiana)
Hesseianaa
```

```python
f_lmbda = sympy.lambdify((t0, t1, t2), Funcion_Costo, 'numpy')
fprime_lmbda = sympy.lambdify((t0, t1, t2), Derivada_Funcion_Costo, 'numpy')
fhess_lmbda = sympy.lambdify((t0, t1, t2), Hesseiana, 'numpy')
```

```python
def func_XY_to_X_Y(f):
    """
    permite tomar dos vectores al mismo tiempo
    """
    return lambda X: np.array(f(X[0], X[1], X[2]))
```

```python
f = func_XY_to_X_Y(f_lmbda)
fprima = func_XY_to_X_Y(fprime_lmbda)
fhessiana = func_XY_to_X_Y(fhess_lmbda)
```

```python
import scipy
from scipy import optimize
x_opt = optimize.fmin_ncg(f, (0, 0, 0), fprime=fprima, fhess=fhessiana)
```
