```python
data = pd.read_csv('parcial2.txt')
data
```

```python
xs = data.drop(columns=['id','lpsa','train'])
```

### Varianza total y componentes principales

```python
# Varianza total
covarianza = xs.cov()
varianza = np.trace(covarianza)
print('Varianza total '+str(round(varianza,3)))
```

```python
# Componentes principales
pca = PCA(len(xs.columns))
pca.fit(xs)
```

```python
print('Componentes principales:')
print(pca.components_)
```

```python
print('Varianza explicada por cada componente:')
print(pca.explained_variance_)
```

```python
print('Porcentaje (%) de varianza explicada:')
print(pca.explained_variance_ratio_.round(3)*100)
```

### Plano de las 2 primeras componentes principales

```python
proy = pca.transform(xs)[:,0:2]
plt.plot(proy[:,0],proy[:,1], 'o')
```

### Regresión de Y sobre los dos primeros componentes principales

```python
model= LinearRegression()
reg = model.fit(proy, data['lpsa'])
```

```python
print('R squared del modelo')
print(reg.score(proy, data['lpsa']))
print('Coeficientes:')
print(reg.coef_)
print('Intercepto:')
print(reg.intercept_)
```

### Regresión sobre la data original

```python
reg = model.fit(xs, data['lpsa'])
```

```python
print('R squared del modelo')
print(reg.score(xs, data['lpsa']))
print('Coeficientes:')
print(reg.coef_)
print('Intercepto:')
print(reg.intercept_)
```

```python
data = pd.read_csv('parcial2_2.txt', delimiter=' ')
data
```

```python
x = data.drop(columns=['Country'])
kmedoides = KMedoids(n_clusters=2, random_state=0, metric='precomputed').fit(x)
```

```python
print('Clasificación:')
print(kmedoides.labels_)
```

```python
adist = data.drop(columns=['Country']).values
amax = np.amax(adist)
adist /= amax

mds = manifold.MDS(n_components=2, dissimilarity="precomputed", random_state=6)
results = mds.fit(adist)

coords = results.embedding_

plt.subplots_adjust(bottom = 0.1)
plt.scatter(
    coords[:, 0], coords[:, 1], marker = 'o'
    )
plt.scatter(
    coords[kmedoides.labels_==1,0], coords[kmedoides.labels_==1,1], marker = 'o'
)
for label, x, y in zip(data.Country, coords[:, 0], coords[:, 1]):
    plt.annotate(
        label,
        xy = (x, y), xytext = (-20, 20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))

plt.show()
```
